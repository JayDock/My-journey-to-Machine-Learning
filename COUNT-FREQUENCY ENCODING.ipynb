## Count or Frequency Encoding

```python
import pandas as pd
```

```python
train_set = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data' , header = None,index_col=None)
train_set.head()
```

```python
columns = [1,3,5,6,7,8,9,13]
```

```python
train_set = train_set[columns]
```

```python
train_set.columns=['Employment','Degree','Status','Designation','family_job','Race','Sex','Country']
```

```python
train_set.head()
```

```python
### Find how many caregories are there in each feature

for feature in train_set.columns[:]:
    print(feature,":",len(train_set[feature].unique()),'labels')   
```

```python
### here country has got the highest labels
```

```python
train_set['Country'].value_counts().to_dict()
```

## Now we are going to replace the countries with the corrsponding number..for example: United stated with 29170, India with 100 and so on

### This process is called mapping

```python
country_map = train_set['Country'].value_counts().to_dict()
```

```python
train_set['Country'] = train_set['Country'].map(country_map)
train_set.head(10)
```

## Advantages
1. Easy to implement
2. Not increaing the feature space

## Disadvantages
1. It will provide the same weight if the frequencues are same
